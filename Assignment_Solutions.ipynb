# Preprocessing dataset

```python
#Identify shape of the dataset
data.shape
```

```python
#Check for missing values
data.isnull().sum()
```

```python
#standardize features
```

```python
# Dropping Unnecessary columns
data = data.drop(["study_resources","student_id"], axis = 1)
```

```python
data.head(2)
```

```python
# use one hot encoder to transform categorical to numerical columns
data_processed = pd.get_dummies(data, columns = [ "gender","socioeconomic_status","family_support","first_generation_student","program_of_study","completed_course"], drop_first = True)
```

```python
data_processed
```

```python
#Defines features and target variable
X_data = data_processed.drop(["final_grade"], axis = 1)
y_data = data_processed["final_grade"]
```

```python
#training the data into training and testing set
from sklearn.model_selection import train_test_split
X_train_data,X_test_data,y_train_data,y_test_data = train_test_split(X_data, y_data, test_size = 0.2, random_state = 42)
```

```python
#scale  the feature
from sklearn.preprocessing import StandardScaler
scaler_data = StandardScaler()
X_train_data_scaled = scaler_data.fit_transform(X_train_data)
X_test_data_scaled = scaler_data.transform(X_test_data)
```

```python
#initiate model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_data,y_train_data)
```

```python
#Model coefficients
print(f"Intercept:{model.intercept_}")
print(f"Coefficent: {model.coef_[0]}")
```

```python
#Predict the value of y
y_pred = model.predict(X_test_data)
```

```python
y_pred
```

```python
#visualize actual and predicted data
plt.scatter(X_test_data, y_test_data, label = "Actual", color = "blue")
plt.plot(X_test_data, y_pred, label = "Predicted" , color = "red")
plt.xlabel("factors")
plt.ylabel("Grade of students")
plt.legend()
plt.show()
```
